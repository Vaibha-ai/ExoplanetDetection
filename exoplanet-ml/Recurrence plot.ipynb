```python
import os
from astropy.io import fits

# Directory containing FITS files
directory = 'astronet/kepler/0114/011442793'

# Get list of FITS files
fits_files = [f for f in os.listdir(directory) if f.endswith('.fits')]

# Loop through each FITS file
for fits_file in fits_files:
    # Open FITS file
    hdul = fits.open(os.path.join(directory, fits_file))
    
    # Get data
    data = hdul[1].data
    
    # Extract flux
    flux = data['PDCSAP_FLUX']  # Use PDCSAP flux for calibrated data
    
    # Save flux values to a text file with the same name as the FITS file
    txt_file_name = os.path.splitext(fits_file)[0] + '.txt'
    with open(txt_file_name, 'w') as f:
        for value in flux:
            f.write(f"{value}\n")
    
    # Close FITS file
    hdul.close()

    print(f"Flux values saved to {txt_file_name}")
```

```python
import os
import pandas as pd
from astropy.io import fits

# Directory containing FITS files
directory = 'astronet/kepler/0114/011442793'

# Get list of FITS files
fits_files = [f for f in os.listdir(directory) if f.endswith('.fits')]

# Create an empty DataFrame to store the flux data
df = pd.DataFrame(columns=['File', 'Flux'])

# Loop through each FITS file
for i, fits_file in enumerate(fits_files):
    # Open FITS file
    hdul = fits.open(os.path.join(directory, fits_file))
    
    # Get data
    data = hdul[1].data
    
    # Extract flux
    flux = data['PDCSAP_FLUX']  # Use PDCSAP flux for calibrated data
    
    # Add flux data as a new row in the DataFrame
    df = df.append({'File': fits_file, 'Flux': flux.tolist()}, ignore_index=True)

# Display the DataFrame
print(df)
```

```python
import os
import pandas as pd
from astropy.io import fits

# Directory containing FITS files
directory = 'astronet/kepler/0114/011442793'

# Get list of FITS files
fits_files = [f for f in os.listdir(directory) if f.endswith('.fits')]

# Create an empty DataFrame to store the flux data
df_non_nan = pd.DataFrame(columns=['File', 'Flux'])

# Loop through each FITS file
for i, fits_file in enumerate(fits_files):
    # Open FITS file
    hdul = fits.open(os.path.join(directory, fits_file))
    
    # Get data
    data = hdul[1].data
    
    # Extract flux
    flux = data['PDCSAP_FLUX']  # Use PDCSAP flux for calibrated data
    
    # Filter out NaN values
    flux_non_nan = [value for value in flux if not pd.isna(value)]
    
    # Add non-NaN flux data as a new row in the DataFrame
    df_non_nan = df_non_nan.append({'File': fits_file, 'Flux': flux_non_nan}, ignore_index=True)

# Display the DataFrame
print(df_non_nan)
```
